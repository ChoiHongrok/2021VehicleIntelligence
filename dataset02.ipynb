```python
import imageio

base_path = "../input/offroad-terrain-dataset-for-autonomous-vehicles/"
label_path = base_path + 'ImageLabels/'
image_path = base_path + 'Images/Images/'
sensor_path = base_path + 'SensorData/'
```

# Images

```python
plt.figure(figsize=(20, 10))
i = 1

img_cnt = 0


for day in os.listdir(image_path):
    
    day_path = os.path.join(image_path, day)
    imgs = os.listdir(day_path)
    
    img_cnt += len(imgs)
    
    for img in imgs:
        path = os.path.join(day_path, img)
        cam_img = imageio.imread(path)
        plt.subplot(3, 2, i)
        i+=1
        plt.imshow(cam_img)
        plt.title(day)
        plt.axis('off')

        break
        
plt.tight_layout()
print(f'all images: {img_cnt}')
```

# ImageLabels

```python
labels = os.listdir(label_path)

for label in labels:
    path = os.path.join(label_path, label)
    
    display(pd.read_csv(path))
```
